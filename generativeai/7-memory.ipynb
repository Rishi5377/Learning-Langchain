### Agent With Memory

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
agent = create_agent(
    model="groq:qwen/qwen3-32b",
    checkpointer=InMemorySaver(),  # Use in-memory checkpointing
)
```

```python
### Run with Thread ID and check memory
config = {"configurable": {"thread_id": "thread-1"}}
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

response = agent.invoke({
    "messages": [HumanMessage(content="hi, My name is Rishanth Reddy, i'm currently studying at sreenidhi institute of science and technology, specializing in DS")]},
    config=config
)

response["messages"][-1]
```

```python
### time to check if the agent remembers the previous message

response = agent.invoke({"messages": [HumanMessage(content="what's my name and where do i study?")]}, config=config)
response["messages"][-1]
```

```python
### Run with a different Thread ID and check if the memory is separate

config_1= {"configurable": {"thread_id": "thread-2"}}
```

```python
response = agent.invoke({
    "messages": [HumanMessage(content="hi, My name is Thrishanth Reddy, i'm currently studying at NIT Warangal, specializing in AIML-DS")]},
    config=config_1
)

response["messages"][-1]
```

```python
## Time to check if the first thread's memory and second thread's memory are intact

response_1 = agent.invoke({"messages": [HumanMessage(content="what's my name and where do i study?")]}, config=config)
response_1["messages"][-1]
```

```python
response_2 = agent.invoke({"messages": [HumanMessage(content="what's my name and where do i study?")]}, config=config_1)
response_2["messages"][-1]
```
