## Generative Application Using LLM Models

### 1. GROQ model

#### a. First Method using langchain

```python
from langchain.chat_models import init_chat_model

chat_model = init_chat_model("llama-3.1-8b-instant", model_provider="groq")

chat_model
```

```python

response = chat_model.invoke("What is the capital of France?")
print(response.content)
```

#### b. Using specfic chat model

```python
from langchain_groq import ChatGroq

chat_model = ChatGroq(model="llama-3.1-8b-instant")

chat_model
```

```python
response = chat_model.invoke("What is the capital of India?")
print(response.content)
```

### 2. GEMINI model

```python
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash")

response = model.invoke("What is the capital of India?")

print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

response = model.invoke("Essay in AI..")

print(response.content)
```
