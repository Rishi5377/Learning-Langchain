### Streaming and Batching

#### Streaming

Most models can stream responses, which means you can get partial results as they are generated.
This is especially useful for long responses, as it allows you to start processing the output before the entire response is ready.

```python
chat_model.invoke("Write a 500 word essay on AI")
```

```python
for chunk in chat_model.stream("Write a 500 word essay on AI"):
    print(chunk.content, end="",flush=True)
```

#### Batch 

Batching is a collection of independent calls to the model that are executed together in a single request. This can significantly reduce latency and improve throughput when making multiple calls to the model.

```python
responses = chat_model.batch([
    "What is the capital of France?",
    "What is the capital of India?",
    "What is the capital of USA?"
])

for response in responses:
    print(response.content)
```

```python
responses = chat_model.batch(
    [
    "What is the capital of France?",
    "What is the capital of India?",
    "What is the capital of USA?"
    ],
config={'max_concurrency': 5, # limit to 5 parallel requests 
}
)

for response in responses:
    print(response.content)
```
